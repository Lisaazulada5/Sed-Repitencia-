## Conversion a 0'1 y 1's de variables SI/NO. En el sisben es 1 o 2

```python
dict_si_no = {
    1:1,
    2:0
}
```

```python
lista_si_no = []
for col in sisben.columns:
    if sisben[col].dtype == 'uint8':
        if set(sisben[col].unique()) == {1,2}:
            lista_si_no.append(col)
            print(sisben[col].unique(), col)
```

```python
for x in lista_si_no:
    sisben[x].replace(dict_si_no, inplace = True)
    print(sisben[x].unique())
```

## Cambio de nombres/ tipos de datos

```python
sisben.rename(columns ={'auto1': 'tiene_auto'}, inplace = True)
```

```python
sisben['puntaje_sisben_3'] = sisben['puntaje_sisben_3'].str.replace(",",".").str.replace(" ", "")
sisben['puntaje_sisben_3'] = pd.to_numeric(sisben['puntaje_sisben_3'], errors='coerce')
```

## Hacinamiento

```python
sisben.tdormir.replace({0:1}, inplace = True)
```

```python
sisben['tdormir'].unique()
```

```python
sisben['indice_hacinamiento'] = sisben['tpersona']/(sisben['tdormir'])
```

```python
sisben["nivel_hacinamiento"] = pd.cut(
    sisben["indice_hacinamiento"],
    [0, 2.4, 4.9, 20],
    labels=['Sin_hacinamiento', 'Hacinamiento_medio', 'Hacinamiento_crítico']
)
```

```python
sisben['nivel_hacinamiento'].value_counts()
```

## Todos servicios

```python
def serv_todos(x):
    suma_servicios = x['energia'] + x['alcanta'] + x['gas'] + x['acueduc']
    return 1 if suma_servicios == 4 else 0
```

```python
sisben['todos_servicios_Pub'] = sisben[['energia','alcanta', 'gas','acueduc']].apply(serv_todos, axis = 1)
```
